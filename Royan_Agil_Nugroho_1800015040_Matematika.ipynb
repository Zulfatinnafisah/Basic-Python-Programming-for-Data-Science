## Report no 1 : Jenis budidaya yang paling diminati di Indonesia?

```python
produksibudidayanasional_new = produksibudidayanasional[produksibudidayanasional['NamaIkan']!= 'total']
produksibudidayanasional_new
```

Dari hasil penghitungan tabel, didapat jenis budidaya yang paling diminati adalah Rumput Laut dan produksi budidaya rumput paling banyak di provinsi Sulawesi Selatan pada tahun 2012 dengan volume sebesar 2.104.446.

```python
nasional = produksibudidayanasional_new[produksibudidayanasional_new.NamaIkan == 'rumput laut']
regional = nasional[nasional.NamaProvinsi == 'Sulawesi Selatan']
plt.bar(regional.Tahun,regional.Volume)
plt.plot(regional.Tahun,regional.Volume)
plt.title('Grafik Rumput Laut di Wilayah Sulawesi Selatan dari Tahun 2003/2012')
plt.xlabel('Tahun')
plt.ylabel('Volume')
plt.show()
```

Dari grafik di atas dapat kita ketahui bahwa terjadi peningkatan yang signifikan dari tahun 2004-2012.
